```python
from os import link
import pandas as pd
data = pd.read_csv('data_road.csv')
```

```python
length = len(data)
```

```python
import pandas as pd

news_data = pd.DataFrame(columns=['news_raw'])

for l in range(length-1):
    news_data = news_data.append({'news_raw': data['content'][l]}, ignore_index=True)
```

```python
news_data.shape
```

```python
news_data
```

# Tokenize

```python
import re
def tokenize(text):
  tokens = re.split('\W+',text)
  return tokens
news_data['news_tokenized'] = news_data['news_raw'].apply(lambda x: tokenize(x.lower()))
```

```python
news_data
```

# Remove Stopwords

```python
import nltk
nltk.download('stopwords')

stopwords = nltk.corpus.stopwords.words('english')

def remove_stopwords(news_tokenized):
  text = [word for word in news_tokenized if word not in stopwords]
  return text

news_data['news_nostop'] = news_data['news_tokenized'].apply(lambda x: remove_stopwords(x))
```

```python
news_data['news_nostop']
```

# Vectorize using TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vect = TfidfVectorizer()
x_tfidf = tfidf_vect.fit_transform(news_data['news_raw'])
x_tfidf
```

```python
tfidf_df = pd.DataFrame(x_tfidf.toarray(), columns=tfidf_vect.get_feature_names_out())
tfidf_df.to_csv('TF_IDF')
```
