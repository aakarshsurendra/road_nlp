```python
from os import link
import pandas as pd
data = pd.read_csv('data_road.csv')
```

```python
length = len(data)
```

```python
import pandas as pd

news_data = pd.DataFrame(columns=['news_raw'])

for l in range(length-1):
    news_data = news_data.append({'news_raw': data['content'][l]}, ignore_index=True)
```

```python
news_data.shape
```

```python
news_data
```

# Tokenize

```python
import re
def tokenize(text):
  tokens = re.split('\W+',text)
  return tokens
news_data['news_tokenized'] = news_data['news_raw'].apply(lambda x: tokenize(x.lower()))
```

```python
news_data
```

# Remove Stopwords

```python
import nltk
nltk.download('stopwords')

stopwords = nltk.corpus.stopwords.words('english')

def remove_stopwords(news_tokenized):
  text = [word for word in news_tokenized if word not in stopwords]
  return text

news_data['news_nostop'] = news_data['news_tokenized'].apply(lambda x: remove_stopwords(x))
```

```python
news_data['news_nostop']
```

# Vectorize using TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vect = TfidfVectorizer()
x_tfidf = tfidf_vect.fit_transform(news_data['news_raw'])
x_tfidf
```

```python
tfidf_df = pd.DataFrame(x_tfidf.toarray(), columns=tfidf_vect.get_feature_names_out())
tfidf_df.to_csv('TF_IDF')
```

```python
import pandas as pd
import spacy
from spacy.tokens import DocBin
import ast
```

```python
df = pd.read_csv('/content/01_03_data_annotation_for_named_entity_recognition.csv')
```

```python
df['annotations'] = df['annotations'].apply(ast.literal_eval)
```

```python
df.info()
```

```python
N = 100
df = pd.concat([df] * N, ignore_index=True)
```

```python
df.info()
```

```python
train_data = df['annotations'][:1200].tolist()
test_data = df['annotations'][1200:].tolist()
```

```python
nlp = spacy.blank("en") # load a new spacy model
db = DocBin() # create a DocBin object
```

```python
for text, annot in train_data: # data in spacy tuple
        doc = nlp.make_doc(text) # create doc object from text
        ents = []
        for start, end, label in annot["entities"]: # add character indexes
            span = doc.char_span(start, end, label=label, alignment_mode="contract")
            if span is None:
                print("Skipping entity")
            else:
                ents.append(span)
```

```python
doc.ents = ents # label the text with the ents
db.add(doc)

db.to_disk("./train.spacy") # save the docbin object
```

```python
nlp2 = spacy.blank("en")
db2 = DocBin()
```

```python
for text, annot in test_data: # data in spacy tuple
        doc2 = nlp2.make_doc(text) # create doc object from text
        ents2 = []
        for start, end, label in annot["entities"]: # add character indexes
            span = doc2.char_span(start, end, label=label, alignment_mode="contract")
            if span is None:
                print("Skipping entity")
            else:
                ents2.append(span)
```

```python
doc2.ents = ents2 # label the text with the ents
db2.add(doc2)

db2.to_disk("./test.spacy") # save the docbin object
```
