# LLM OPEN AI EVAUVATIONS

```python
# Load the CSV files
llm_data_test_df = pd.read_csv('llm_test_data_t.csv')
transformed_df = pd.read_csv('input/data_extracted_transformed_openai.csv')


# Define the columns to compare
columns_to_compare = ['age', 'vehicle_type', 'reason', 'fatalities', 'injured', 'gender', 'time']

# Initialize the accuracy score dictionary
accuracy_scores = {}
correct_counts = {}
incorrect_counts = {}

# Iterate over the columns to compare
for column in columns_to_compare:
    # Initialize the accuracy score for the current column
    accuracy_score = 0
    correct_count = 0
    incorrect_count = 0
    
    # Iterate over the rows in the llm_data_test dataframe
    for index, row in llm_data_test_df.iterrows():
        # Get the ID from the current row
        id = row['id']
        
        # Filter the transformed dataframe to get the row with the same ID
        transformed_row = transformed_df[transformed_df['id'] == id]
        
        # Check if the transformed_row is not empty
        if not transformed_row.empty:
            # Get the value from the transformed dataframe
            transformed_value = transformed_row[column].values
            if len(transformed_value) > 0:
                transformed_value = transformed_value[0]
                
                # If the column value is a list, convert it to a string for comparison
                if isinstance(transformed_value, list):
                    transformed_value = ', '.join(transformed_value)
                else:
                    transformed_value = str(transformed_value)
                
                # Get the value from the llm_data_test dataframe
                llm_data_test_value = row[column]
                
                # Check if the values match
                if transformed_value.lower() == str(llm_data_test_value).lower():
                    accuracy_score += 1
                    correct_count += 1
                else:
                    incorrect_count += 1
    
    # Calculate the accuracy score for the current column
    accuracy_scores[column] = accuracy_score / len(llm_data_test_df)
    correct_counts[column] = correct_count
    incorrect_counts[column] = incorrect_count

# Print the accuracy scores for each entity
for column, accuracy_score in accuracy_scores.items():
    print(f'Accuracy score for {column}: {accuracy_score}')
    
# Print the counts for correct and incorrect matches for each entity
for column in columns_to_compare:
    print(f'Correct count for {column}: {correct_counts[column]}')
    print(f'Incorrect count for {column}: {incorrect_counts[column]}')
```
